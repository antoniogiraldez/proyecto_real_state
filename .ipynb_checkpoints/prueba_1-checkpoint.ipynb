```python
import pandas as pd
import io
import seaborn as sns
```

```python
df=pd.read_csv('LArealestate.csv')
df
```

```python
df.isna().sum()
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
import matplotlib.pyplot as plt
plt.figure()
sns.scatterplot(x=df['beds'],y=df['price'])
plt.figure()
sns.scatterplot(x=df['baths'],y=df['price'])
plt.figure()
sns.scatterplot(x=df['sqft'],y=df['price'])
plt.show()
```

```python
sns.catplot(x='city',y='price',data=df)
plt.show()
```

```python
for i in df['city'].unique():
    plt.figure()
    sns.boxplot(y=df[df['city']==i]['price'])
    plt.show()
```
