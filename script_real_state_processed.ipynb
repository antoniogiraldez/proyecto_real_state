```python
df['city'] = df.city.replace('Culver City', 'culver city')
```

```python
df['real_state'] = [df['date'][i][10:] for i in range(len(df.date))]
```

```python
df['date'] = [df['date'][i][:10] for i in range(len(df.date))]
```

```python
df_num = df[['beds', 'baths', 'sqft', 'price']]
df_num
```

```python
df_cat = df[['address','date','city', 'type', 'real_state']]
df_cat
```

```python
variable_type = pd.get_dummies(df_cat['type'], drop_first = True)
df_cat['type'] = variable_type.values
df_cat
```

```python
variable_city = pd.get_dummies(df_cat['city'], drop_first = True)
df1 = pd.concat([df_num, variable_city], axis = 1)
```

```python
df_cat['date'] = pd.to_datetime(df['date'])
```

```python
df_cat['date'] = df_cat['date'].dt.year
```

```python
variable_date = pd.get_dummies(df_cat['date'], drop_first = True)
df1 = pd.concat([df1, variable_date], axis = 1)
```

```python
variable_rs = pd.get_dummies(df_cat['real_state'], drop_first = True)
df1 = pd.concat([df1, variable_rs], axis = 1)
```
